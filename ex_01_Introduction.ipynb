## Different Cell Types

For an auto-grading jupyter notebook, there are three types of cells:
- **Read-Only.** Generally all markdown cells and some code cells fall into this category. These cells are annotated with a special "read only" flag and _should not_ be edited under any circumstances. Any changes that are made in these cells will be _discarded_ during the auto-grading process. Read-only code cells can be identified by a special "DO NOT MODIFY" comment at the start.
- **Answers.** These are the only cells that should be edited while working on the exercise. Answer cells usually contain clear comments what kind of code should be added at which points to identify the exercise.
- **Tests.** These are a special case of the *read-only* cell. Test cells contain a series of assert statements that will test the validity and correctness of a previously implemented *answer* cell. For instance, if the goal of a task was to compute a specific value, then the corresponding test cell will check if the variable contains a value of the correct type and within an approximately correct value range, for example.

### Test Assertions

As previously introduced, testing cells contain various assertions which check the validity and the correctness of an implemented answer. Each test cell is associated with a number of *points* &mdash; if a test cell is able to execute *without any exception* then those points are awarded to the final score.

However, the assertions that are visible in the published version of the notebook only constitute the **public tests**. The public tests are a subset of the tests that are supposed to mainly check for the technical validity of the solution (correct data type or shape of tensors) and sometimes offer hints toward the correctness (approximately correct value range). Passing the public tests is a good indication regarding the correctness of the exercise, but *does not guarantee* the points!

In addition to the public tests, each testing cell also contains a set of **hidden tests** which are only applied during grading. Only if both the public *and* the hidden tests are passed, the points of the cell are awarded to the final score.
